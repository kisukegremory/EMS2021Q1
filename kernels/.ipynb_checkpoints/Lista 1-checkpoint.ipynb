# Questão 6) 
---
(Adaptado das listas do prof. João Moreira – UABC). Considere que um país sem gás natural descubra uma
província de petróleo que contenha gás natural associado e durante 30 anos a taxa de descoberta de novas reservas
de gás natural seja constante e equivalha a 15 bilhões de $m^3$
/ano. Após 30 anos, a taxa de descoberta de novas
reservas cai drasticamente e pode ser considerada nula. Determine a evolução das reservas de gás natural deste
país ao longo de 50 anos considerando que o coeficiente de extração de gás natural, λ, seja de 0,08 $ano^{-1}$
. 

Se anos <= 30:
$$
X_1(t) = \frac{F_0}{\lambda}(1 - e^{-\lambda t})\\
X_1(t) = \frac{15}{0.08}(1 - e^{-0.08 t})\\
$$
Se anos > 30:
$$
X_2(t) = X_1(30)(e^{-\lambda (t-30)})\\
X_2(t) = X_1(30)(e^{-0.08 (t-30)})\\
$$

```python
plt.figure(figsize=(16,3))
x1 = np.arange(0,31,1)
y1 = 15/0.08 * (1 - np.exp(-0.08*x1))
plt.plot(x1,y1, label='Durante descobertas')

x2 = np.arange(30,51,0.1)
y2 = y1[-1] * (np.exp(-0.08*(x2-x1[-1])))
plt.plot(x2,y2, label='Fim das descobertas')
plt.title("Questão 6",fontsize=18)
plt.legend()
plt.ylabel("Bilhões de m³")
plt.xlabel("Tempo")
plt.grid(alpha=0.4)
plt.show()
```

# Questão 8) 
---
(Adaptado das listas do prof. João Moreira – UABC). Suponha que seja possível modelar o comportamento
temporal do crescimento natural da floresta amazônica segundo a função logística, isto é:

$$
X' = rX(1 - \frac{X}{X_s})
$$


onde r e Xs são os parâmetros descritivos importantes. Estes parâmetros podem ser determinados a partir de
algumas observações da evolução temporal da área florestal amazônica. Por exemplo, a taxa de recuperação
natural (regeneração) foi estimada em 2010 em 1200 $km^2$
/ano; supõe-se que a floresta cobria uma área de
aproximadamente 8 milhões de $km^2$
em 1500 e que em 2010 cerca de 20 % de sua área original já tinha sido
desmatada. Determine, a partir dos dados acima:  
a) os parâmetros r e Xs para a floresta amazônica. Dica: tome t = 2010 e considere que em 1500 a floresta estava
intocada e em sua área máxima.  
b) a evolução da área florestal amazônica em função do tempo X(t), a partir de 2010, considerando que não haverá
qualquer tipo de extração.  
c) Determine a área florestal em 2100.

a) 

$$
X'(2010) = rX(2010)(1 - \frac{X(2010)}{X_s})\\
1200 = r0.8*8.10^6(1 - 0.8)\\
r = \frac{1200}{0.8*8*10^6(0.2)} = 0.0009375/ano\\
X_s = 8*10^6 km^2\\
$$

b-)
$$
X' = rX(1 - \frac{X}{X_s})\\
\frac{dX}{X(1 - \frac{X}{X_s})} = rdt\\
dX[\frac{1}{X} + \frac{1}{X_s - X}] = rdt\\
ln(X) - ln(X_s - X)|^{t=t}_{t=0} = rt\\
ln(X_s - X) - ln(X) |^{t=t}_{t=0} = -rt\\
ln(\frac{X_s - X(t)}{X_s - X(0)}) - ln(\frac{X(t)}{X(0)}) = -rt\\
ln(\frac{X_s - X(t)}{X_s - X(0)}\frac{X(0)}{X(t)}) = -rt\\
\frac{X_s - X(t)}{X_s - X(0)}\frac{X(0)}{X(t)} = e^{-rt}\\
\frac{X_s - X(t)}{X(t)}= \frac{X_s - X(0)}{X(0)}e^{-rt}\\
\frac{X_s}{X(t)} - 1 = \frac{X_s - X(0)}{X(0)}e^{-rt}\\
\frac{X_s}{X(t)} = 1 + \frac{X_s - X(0)}{X(0)}e^{-rt}\\
X(t) = \frac{X_s}{1 + \frac{X_s - X(0)}{X(0)}e^{-rt}} \\
$$

```python

x_s = 8*(10**6)
x_0 = 0.8*8*(10**6)
r = 0.0009375
t = np.arange(0,8000,1)

anos = np.arange(2010,2010+t.shape[0],1)
y = x_s/(1+ ((x_s - x_0)/x_0)*np.exp(-r*t))
plt.figure(figsize=(16,5))
plt.plot(anos,y,label='evolução da área florestal amazônica')
plt.scatter(2100,y[2100-2010],s=100,color='orange',label=f'ano 2100 equivalente a {y[2100-2010].round()}km²')
plt.legend()
plt.ylabel("área florestal amazônica km^2")
plt.xlabel("Anos")
plt.grid(alpha=0.4)
plt.show()
```

# Questão 9)
--- 
Considere que no problema anterior há uma taxa de extração proporcional à quantidade de área florestal
existente, com um coeficiente de extração $\lambda e = 0,0005 ano^{-1}$.  
Determine:  
a) O valor de equilíbrio da quantidade de reservas  
b) A evolução temporal da quantidade de reservas em função do tempo, X(t)  


$$
X(t) = \frac{x_0 e^{(r-\lambda_e)t}}{1 + \frac{r}{r - \lambda_e}x_0 (e^{(r-\lambda_e)t} - 1) }
$$

Valor de Equilibrio: 

$$
X_{eq} = \frac{r - \lambda_e}{r}X_s\\
X_{eq} = \frac{0.0009375 - 0.0005}{0.0009375}8*10^6\\
X_{eq} = 3,73*10^6
$$

Tentativa de integral:

$$
\int_{X_0}^X \frac{1}{X'}dX' + \int_{X_0}^X \frac{a}{1-aX'}dX' = \int_{0}^t b dt'
$$
$$
u = (1-aX')\\
du = -adx\\
\int_{u_0}^u \frac{-1}{u}du\\
-ln(u)|_{1-aX_0}^{1-aX}
$$
$$
ln(X)|_{X_0}^{X} -ln(u)|_{1-aX_0}^{1-aX} = bt \\
ln(\frac{X}{X_0}) + ln(\frac{1-aX_0}{1-aX}) = bt \\
ln(\frac{X}{X_0}*\frac{1-aX_0}{1-aX}) = bt \\
X*\frac{1-aX_0}{1-aX} = X_0e^{bt} \\
\frac{X}{1-aX} = \frac{X_0e^{bt}}{1-aX_0} \\
\frac{1-aX}{X} = \frac{1-aX_0}{X_0e^{bt}} \\
\frac{1}{X} - a = \frac{1-aX_0}{X_0e^{bt}} \\
\frac{1}{X} = \frac{1-aX_0}{X_0e^{bt}} + a \\
\frac{1}{X} = \frac{1-aX_0 + a*X_0e^{bt}}{X_0e^{bt}}\\
\frac{1}{X} = \frac{1 + a(-X_0 + X_0e^{bt})}{X_0e^{bt}}\\
X = \frac{X_0e^{bt}}{1 + aX_0(e^{bt} - 1)}\\
$$

Substituindo os valores:


$$
a = (\frac{r}{(r-\lambda_e)X_s})\\
b = (r-\lambda_e)\\
X = \frac{X_0e^{bt}}{1 + aX_0(e^{bt} - 1)}\\
X = \frac{X_0e^{(r-\lambda_e)t}}{1 + (\frac{r}{(r-\lambda_e)X_s})X_0(e^{bt} - 1)}\\
$$

```python
x_0 = 0.8*8*(10**6)
t = np.arange(0,5000,1)
anos = np.arange(2010,2010+t.shape[0],1)
y = (x_0*np.exp((0.0009375 - 0.0005)*t)) #/(1 + ((0.0009375*x_0)/(0.0009375 - 0.0005))*(np.exp(t*(0.0009375 - 0.0005)) - 1))
y = y/(1 + x_0*(0.0009375)/(0.0009375 - 0.0005)*(np.exp(t*(0.0009375 - 0.0005)) - 1)/x_s)
plt.figure(figsize=(16,5))
plt.plot(anos,y,label='evolução da área florestal amazônica')
plt.scatter(2100,y[2100-2010],s=100,color='r',label=f'ano 2100 equivalente a {y[2100-2010].round()}km²')
plt.legend()
plt.ylabel("área florestal amazônica km^2")
plt.xlabel("Anos")
plt.grid(alpha=0.4)
plt.show()
```

# Questão 11
--- 
Considere que atualmente a taxa de emissão de CO2, um dos gases do efeito estufa, seja de 4,5 ppm/ano e que
a capacidade de assimilação do planeta (remoção desses gases) possa ser representada por uma taxa percentual
de 1 %/ano.  
a) Se em 2000 a concentração de CO2 era de 375 ppm, calcule esta concentração em 2060.  
b) Se o planeta não contribuísse com a remoção de CO2, qual seria a concentração em 2060?  
Respostas: a) 395 ppm; b) 430 ppm.

a)
$$
X_1(t) = X_{0}e^{-\lambda t} + \frac{F_0}{\lambda}(1 - e^{-\lambda t})\\
X_1(t) = 375(e^{-0.01 t}) + \frac{4.5}{0.01}(1 - e^{-0.01 t})\\
$$
b)
$$
X_1(t) = 375 + 4.5t\\
$$

```python
plt.figure(figsize=(16,3))
t = np.arange(0,100,1)
anos = np.arange(2000,2000+t.shape[0],1)

y1 =  375*(np.exp(-0.01*t)) + 4.5/0.01 * (1 - np.exp(-0.01*t))
plt.plot(anos,y1, label='Com contribuição do planeta')
plt.scatter(2060,y1[60],s=100,label=f'Com Contribuição em 2060 : {y1[60].round(2)} ppm')
y2 = 375 + 4.5*t
plt.plot(anos,y2, label='Sem contribuição do planeta')
plt.scatter(2060,y2[60],s=100,label=f'Sem Contribuição em 2060 : {y2[60]} ppm')
plt.title("Questão 11",fontsize=18)
plt.legend()
plt.ylabel("ppm")
plt.xlabel("Tempo")
plt.grid(alpha=0.4)
plt.show()
```

# Questão 12
---
Suponha que a concentração de CO2 na atmosfera seja de 375 ppm e que a fração f das emissões terrestres
que retornam à superfície devido ao efeito estufa seja 0,397. Estime o incremento na temperatura média da
superfície terrestre se a concentração de CO2 na atmosfera crescer para 395 ppm. Dica: use teoria da perturbação.
Resposta: ΔT = 2,66 K. 


Estado 1:  
C_1 = 375  
f_1 = 0,397  
Estado 2:  
C_2 = 395  
f_2 = $\frac{395*0,397}{375} = 0.418 $  

portanto:  

q_1 = $C_1*f_1$ = 154.83  
q_2 = $C_2*f_2$ = 163.02

$\Delta q = 8.2$

portanto:

$ q_f = \Delta q + q_0 = 8.2 + 390 = 398.2 $  

$$ T_2 = (\frac{398.2}{5.67.10^{-8}})^{1/4} - 273.2 = 16.3 $$
$$ \Delta T = 16.3 - 14.7 = 1.6^{\circ}C $$

```python
C_1 = 375 
f_1 = 0.397
C_2 = 395
w = 390
f_2 = round(C_2*f_1/C_1,3)
q_1 = round(w*f_1,3) 
q_2 = round(w*f_2,3)
q_f = round(q_2-q_1,3)
T_2 = round(((w+q_f)/(5.67e-08))**(1/4) - 273.2,2)
T_1 = 14.7
T_f = round(T_2 - T_1,2)
print(f_2,q_1,q_2,q_f,T_2,T_f)
```
