## Quest√£o 3
---

Ap√≥s uma interven√ß√£o governamental, em 2018, uma floresta que vinha sofrendo uma agressiva
degrada√ß√£o devido √† interven√ß√£o humana (extra√ß√£o) deixou de ser desmatada. A partir de 2018, a √°rea da floresta
em fun√ß√£o do tempo passou a apresentar uma taxa de crescimento natural, ùêπ(ùë°), cujo comportamento temporal
pode ser descrito pela fun√ß√£o log√≠stica, isto √©:
$$
X' = rX(1 - \frac{X}{X_s})
$$

onde ùëã(ùë°) √© a √°rea da floresta, em km2
, ùëü √© o coeficiente de reposi√ß√£o natural da floresta, expresso em $ùëéùëõùëú^{‚àí1}$
, e
$ùëã_ùëÜ$ √© a capacidade de suporte da floresta, sendo ùêπ(ùë°) expressa em km2
/ano. Em 2010, a taxa de crescimento natural
da floresta foi estimada em 10000 km2
/ano e a √°rea da floresta foi estimada 4 milh√µes de km2
, o que corresponde
a 80% da √°rea que ela apresentava antes de qualquer interven√ß√£o humana (extra√ß√£o). Com base nessas
informa√ß√µes, determine:  
a) a √°rea da floresta em fun√ß√£o do tempo, ùëã(ùë°), a partir de 2010.  
b) a √°rea da floresta em 2050.  

a) 

$$
X'(2010) = rX(2010)(1 - \frac{X(2010)}{X_s})\\
10000 = r4.10^6(1 - 0.8)\\
r = \frac{10000}{4*10^6(0.2)} = 0.0125/ano\\
X_s = 5*10^6 km^2\\
$$

b-)
$$
X' = rX(1 - \frac{X}{X_s})\\
\frac{dX}{X(1 - \frac{X}{X_s})} = rdt\\
dX[\frac{1}{X} + \frac{1}{X_s - X}] = rdt\\
ln(X) - ln(X_s - X)|^{t=t}_{t=0} = rt\\
ln(X_s - X) - ln(X) |^{t=t}_{t=0} = -rt\\
ln(\frac{X_s - X(t)}{X_s - X(0)}) - ln(\frac{X(t)}{X(0)}) = -rt\\
ln(\frac{X_s - X(t)}{X_s - X(0)}\frac{X(0)}{X(t)}) = -rt\\
\frac{X_s - X(t)}{X_s - X(0)}\frac{X(0)}{X(t)} = e^{-rt}\\
\frac{X_s - X(t)}{X(t)}= \frac{X_s - X(0)}{X(0)}e^{-rt}\\
\frac{X_s}{X(t)} - 1 = \frac{X_s - X(0)}{X(0)}e^{-rt}\\
\frac{X_s}{X(t)} = 1 + \frac{X_s - X(0)}{X(0)}e^{-rt}\\
X(t) = \frac{X_s}{1 + \frac{X_s - X(0)}{X(0)}e^{-rt}} \\
X(t) = \frac{5*10^6}{1 + 0.25e^{-0.0125t}} \\
$$

```python
x_s = 5e6
x_0 = 0.8*x_s
r = 0.0125
t = np.arange(0,500,1)

anos = np.arange(2010,2010+t.shape[0],1)
y = x_s/(1+ ((x_s - x_0)/x_0)*np.exp(-r*t))
plt.figure(figsize=(16,5))
plt.plot(anos,y,label='evolu√ß√£o da √°rea florestal amaz√¥nica')
plt.scatter(2050,y[2050-2010],s=200,color='orange',label=f'ano 2050 equivalente a {y[2050-2010].round()}km¬≤')
plt.legend()
plt.ylabel("√°rea florestal amaz√¥nica km^2")
plt.xlabel("Anos")
plt.grid(alpha=0.4)
plt.show()
```

## Quest√£o 4
--- 

Suponha que a concentra√ß√£o atual de CO2 na atmosfera seja de 375 ppm e que a fra√ß√£o das emiss√µes terrestres
que retornam √† superf√≠cie devido ao efeito estufa seja 0,397, para uma emissividade da superf√≠cie da Terra de 390
W/m2
. Estime o incremento na temperatura m√©dia da superf√≠cie terrestre se a concentra√ß√£o de CO2 na atmosfera
crescer para 425 ppm. Considere a temperatura m√©dia atual da superf√≠cie da Terra de 14,7 ¬∞C.

Estado 1:  
C_1 = 375  
f_1 = 0,45 
Estado 2:  
C_2 = 425  
f_2 = $\frac{425*0,397}{375} = 0.45 $  

portanto:  

q_1 = $C_1*f_1$ = 154.83  
q_2 = $C_2*f_2$ = 175.5

$\Delta q = 20.67$

portanto:

$ q_f = \Delta q + q_0 = 20.67 + 390 = 410.67 $  

$$ T_2 = (\frac{410.67}{5.67.10^{-8}})^{1/4} - 273.2 = 18.53 $$
$$ \Delta T = 18.53 - 14.7 = 3.83^{\circ}C $$

```python
C_1 = 375 
f_1 = 0.397
C_2 = 425
w = 390
f_2 = round(C_2*f_1/C_1,3)
q_1 = round(w*f_1,3) 
q_2 = round(w*f_2,3)
q_f = round(q_2-q_1,3)
T_2 = round(((w+q_f)/(5.67e-08))**(1/4) - 273.2,2)
T_1 = 14.7
T_f = round(T_2 - T_1,2)
print(f_2,q_1,q_2,q_f,T_2,T_f)
```
