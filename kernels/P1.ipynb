## Questão 3
---

Após uma intervenção governamental, em 2018, uma floresta que vinha sofrendo uma agressiva
degradação devido à intervenção humana (extração) deixou de ser desmatada. A partir de 2018, a área da floresta
em função do tempo passou a apresentar uma taxa de crescimento natural, 𝐹(𝑡), cujo comportamento temporal
pode ser descrito pela função logística, isto é:
$$
X' = rX(1 - \frac{X}{X_s})
$$

onde 𝑋(𝑡) é a área da floresta, em km2
, 𝑟 é o coeficiente de reposição natural da floresta, expresso em $𝑎𝑛𝑜^{−1}$
, e
$𝑋_𝑆$ é a capacidade de suporte da floresta, sendo 𝐹(𝑡) expressa em km2
/ano. Em 2010, a taxa de crescimento natural
da floresta foi estimada em 10000 km2
/ano e a área da floresta foi estimada 4 milhões de km2
, o que corresponde
a 80% da área que ela apresentava antes de qualquer intervenção humana (extração). Com base nessas
informações, determine:  
a) a área da floresta em função do tempo, 𝑋(𝑡), a partir de 2010.  
b) a área da floresta em 2050.  

a) 

$$
X'(2010) = rX(2010)(1 - \frac{X(2010)}{X_s})\\
10000 = r4.10^6(1 - 0.8)\\
r = \frac{10000}{4*10^6(0.2)} = 0.0125/ano\\
X_s = 5*10^6 km^2\\
$$

b-)
$$
X' = rX(1 - \frac{X}{X_s})\\
\frac{dX}{X(1 - \frac{X}{X_s})} = rdt\\
dX[\frac{1}{X} + \frac{1}{X_s - X}] = rdt\\
ln(X) - ln(X_s - X)|^{t=t}_{t=0} = rt\\
ln(X_s - X) - ln(X) |^{t=t}_{t=0} = -rt\\
ln(\frac{X_s - X(t)}{X_s - X(0)}) - ln(\frac{X(t)}{X(0)}) = -rt\\
ln(\frac{X_s - X(t)}{X_s - X(0)}\frac{X(0)}{X(t)}) = -rt\\
\frac{X_s - X(t)}{X_s - X(0)}\frac{X(0)}{X(t)} = e^{-rt}\\
\frac{X_s - X(t)}{X(t)}= \frac{X_s - X(0)}{X(0)}e^{-rt}\\
\frac{X_s}{X(t)} - 1 = \frac{X_s - X(0)}{X(0)}e^{-rt}\\
\frac{X_s}{X(t)} = 1 + \frac{X_s - X(0)}{X(0)}e^{-rt}\\
X(t) = \frac{X_s}{1 + \frac{X_s - X(0)}{X(0)}e^{-rt}} \\
X(t) = \frac{5*10^6}{1 + 0.25e^{-0.0125t}} \\
$$

```python
x_s = 5e6
x_0 = 0.8*x_s
r = 0.0125
t = np.arange(0,500,1)

anos = np.arange(2010,2010+t.shape[0],1)
y = x_s/(1+ ((x_s - x_0)/x_0)*np.exp(-r*t))
plt.figure(figsize=(16,5))
plt.plot(anos,y,label='evolução da área florestal amazônica')
plt.scatter(2050,y[2050-2010],s=200,color='orange',label=f'ano 2050 equivalente a {y[2050-2010].round()}km²')
plt.legend()
plt.ylabel("área florestal amazônica km^2")
plt.xlabel("Anos")
plt.grid(alpha=0.4)
plt.show()
```

Suponha que a concentração atual de CO2 na atmosfera seja de 375 ppm e que a fração das emissões terrestres
que retornam à superfície devido ao efeito estufa seja 0,397, para uma emissividade da superfície da Terra de 390
W/m2
. Estime o incremento na temperatura média da superfície terrestre se a concentração de CO2 na atmosfera
crescer para 425 ppm. Considere a temperatura média atual da superfície da Terra de 14,7 °C.

Estado 1:  
C_1 = 375  
f_1 = 0,45 
Estado 2:  
C_2 = 425  
f_2 = $\frac{425*0,397}{375} = 0.45 $  

portanto:  

q_1 = $C_1*f_1$ = 154.83  
q_2 = $C_2*f_2$ = 175.5

$\Delta q = 20.67$

portanto:

$ q_f = \Delta q + q_0 = 20.67 + 390 = 410.67 $  

$$ T_2 = (\frac{410.67}{5.67.10^{-8}})^{1/4} - 273.2 = 18.53 $$
$$ \Delta T = 18.53 - 14.7 = 3.83^{\circ}C $$

```python
C_1 = 375 
f_1 = 0.397
C_2 = 425
w = 390
f_2 = round(C_2*f_1/C_1,3)
q_1 = round(w*f_1,3) 
q_2 = round(w*f_2,3)
q_f = round(q_2-q_1,3)
T_2 = round(((w+q_f)/(5.67e-08))**(1/4) - 273.2,2)
T_1 = 14.7
T_f = round(T_2 - T_1,2)
print(f_2,q_1,q_2,q_f,T_2,T_f)
```

## Questão 8) 
---
(Adaptado das listas do prof. João Moreira – UABC). Suponha que seja possível modelar o comportamento
temporal do crescimento natural da floresta amazônica segundo a função logística, isto é:

$$
X' = rX(1 - \frac{X}{X_s})
$$


onde r e Xs são os parâmetros descritivos importantes. Estes parâmetros podem ser determinados a partir de
algumas observações da evolução temporal da área florestal amazônica. Por exemplo, a taxa de recuperação
natural (regeneração) foi estimada em 2010 em 1200 $km^2$
/ano; supõe-se que a floresta cobria uma área de
aproximadamente 8 milhões de $km^2$
em 1500 e que em 2010 cerca de 20 % de sua área original já tinha sido
desmatada. Determine, a partir dos dados acima:  
a) os parâmetros r e Xs para a floresta amazônica. Dica: tome t = 2010 e considere que em 1500 a floresta estava
intocada e em sua área máxima.  
b) a evolução da área florestal amazônica em função do tempo X(t), a partir de 2010, considerando que não haverá
qualquer tipo de extração.  
c) Determine a área florestal em 2100.

a) 

$$
X'(2010) = rX(2010)(1 - \frac{X(2010)}{X_s})\\
1200 = r0.8*8.10^6(1 - 0.8)\\
r = \frac{1200}{0.8*8*10^6(0.2)} = 0.0009375/ano\\
X_s = 8*10^6 km^2\\
$$

b-)
$$
X' = rX(1 - \frac{X}{X_s})\\
\frac{dX}{X(1 - \frac{X}{X_s})} = rdt\\
dX[\frac{1}{X} + \frac{1}{X_s - X}] = rdt\\
ln(X) - ln(X_s - X)|^{t=t}_{t=0} = rt\\
ln(X_s - X) - ln(X) |^{t=t}_{t=0} = -rt\\
ln(\frac{X_s - X(t)}{X_s - X(0)}) - ln(\frac{X(t)}{X(0)}) = -rt\\
ln(\frac{X_s - X(t)}{X_s - X(0)}\frac{X(0)}{X(t)}) = -rt\\
\frac{X_s - X(t)}{X_s - X(0)}\frac{X(0)}{X(t)} = e^{-rt}\\
\frac{X_s - X(t)}{X(t)}= \frac{X_s - X(0)}{X(0)}e^{-rt}\\
\frac{X_s}{X(t)} - 1 = \frac{X_s - X(0)}{X(0)}e^{-rt}\\
\frac{X_s}{X(t)} = 1 + \frac{X_s - X(0)}{X(0)}e^{-rt}\\
X(t) = \frac{X_s}{1 + \frac{X_s - X(0)}{X(0)}e^{-rt}} \\
$$

```python

x_s = 8*(10**6)
x_0 = 0.8*8*(10**6)
r = 0.0009375
t = np.arange(0,8000,1)

anos = np.arange(2010,2010+t.shape[0],1)
y = x_s/(1+ ((x_s - x_0)/x_0)*np.exp(-r*t))
plt.figure(figsize=(16,5))
plt.plot(anos,y,label='evolução da área florestal amazônica')
plt.axhline(y[2100-2010],color='orange',label=f'ano 2100 equivalente a {y[2100-2010].round()}km²')
plt.legend()
plt.ylabel("área florestal amazônica km^2")
plt.xlabel("Anos")
plt.grid(alpha=0.4)
plt.show()
```
