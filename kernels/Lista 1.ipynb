(Adaptado das listas do prof. Jo칚o Moreira  UABC). Suponha que seja poss칤vel modelar o comportamento
temporal do crescimento natural da floresta amaz칪nica segundo a fun칞칚o log칤stica, isto 칠:

$$

onde r e Xs s칚o os par칙metros descritivos importantes. Estes par칙metros podem ser determinados a partir de
algumas observa칞칫es da evolu칞칚o temporal da 치rea florestal amaz칪nica. Por exemplo, a taxa de recupera칞칚o

qualquer tipo de extra칞칚o.  
c) Determine a 치rea florestal em 2100.

a) 

$$
X'(2010) = rX(2010)(1 - \frac{X(2010)}{X_s})\\
1200 = r0.8*8.10^6(1 - 0.8)\\
r = \frac{1200}{0.8*8*10^6(0.2)} = 0.0009375/ano\\

$$

b-)
$$
X' = rX(1 - \frac{X}{X_s})\\
\frac{dX}{X(1 - \frac{X}{X_s})} = rdt\\
dX[\frac{1}{X} + \frac{1}{X_s - X}] = rdt\\
ln(X) - ln(X_s - X)|^{t=t}_{t=0} = rt\\
ln(X_s - X) - ln(X) |^{t=t}_{t=0} = -rt\\
ln(\frac{X_s - X(t)}{X_s - X(0)}) - ln(\frac{X(t)}{X(0)}) = -rt\\
ln(\frac{X_s - X(t)}{X_s - X(0)}\frac{X(0)}{X(t)}) = -rt\\

\frac{X_s}{X(t)} - 1 = \frac{X_s - X(0)}{X(0)}e^{-rt}\\
\frac{X_s}{X(t)} = 1 + \frac{X_s - X(0)}{X(0)}e^{-rt}\\
